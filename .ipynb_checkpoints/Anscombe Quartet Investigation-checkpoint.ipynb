### 2. Background of Anscombe Quartet Dataset

Frank Anscombe was an English statician, born in 1918. Frank Anscombe died in 2001 but throughout his life he made numerous contributions to the field of statistics. One such contribution is the use of 4 simple datasets to demonstrate the dangers of soley using summary statistics to analyse data, in a scientific paper he published in the "The American Statistician, Vol. 27, No. 1. (Feb., 1973)". These 4 datasets since come to be known as "Anscombe's Quartet dataset". In not so many words, Anscombe's lesson to the world was that, while summary statistics can be very useful in statistical analysis, they can also be misleading which is why it is so important to visualise data in tandem with summary statistics. Both will contribute to the understanding of the data. The Anascombe Quartet dataset, as the name suggests, consists of 4 datasets which can be seen in the figure below.

```python
# Magic cell code to fix possible bugs with using matplotlib in jupyter notebook
%matplotlib inline 

#Importing Matplotlib and Numpy packages
import matplotlib.pyplot as plt
import numpy as np

# Imports the HTML and Display package for IPython
from IPython.display import HTML 
from IPython.display import display

# Code adapted from https://stackoverflow.com/questions/31517194/how-to-hide-one-specific-cell-input-or-output-in-ipython-notebook
# Displays the link and associated text, if the link is clicked the javascript function code_toggle runs and either hides or shows the contents of the jupyter notebook cell.
tag = HTML('''<script> 
code_show=true; 
function code_toggle() {
    if (code_show){
        $('div.cell.code_cell.rendered.selected div.input').hide();
    } else {
        $('div.cell.code_cell.rendered.selected div.input').show();
    }
    code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
To show/hide this cell's raw code input, click <a href="javascript:code_toggle()">here</a>. This cell contains the code for plotting the 4 graphs for Anscombe's Quartet Dataset ''')
display(tag) 

# Changes all arguments of the array to type float adapted from https://stackoverflow.com/questions/28393103/typeerror-cannot-perform-reduce-with-flexible-type
x1 = np.array((XdatasetI)).astype(np.float)
y1 = np.array(((YdatasetI))).astype(np.float)
x2 = np.array((XdatasetII)).astype(np.float)
y2 = np.array((YdatasetII)).astype(np.float)
x3 = np.array((XdatasetIII)).astype(np.float)
y3 = np.array((YdatasetIII)).astype(np.float)
x4 = np.array((XdatasetIV)).astype(np.float)
y4 = np.array((YdatasetIV)).astype(np.float)

# (n) length of the 4 Datasets
Z1 = len(y1)
Z2 = len(y2)
Z3 = len(y3)
Z4 = len(y4)

# Mean of the x's and mean of the y's for each dataset
meanx1 = (sum(x1)/len(x1))
meany1 = (sum(y1)/len(y1))
meanx2 = (sum(x2)/len(x2))
meany2 = (sum(y2)/len(y2))
meanx3 = (sum(x3)/len(x3))
meany3 = (sum(y3)/len(y3))
meanx4 = (sum(x4)/len(x4))
meany4 = (sum(y4)/len(y4))

# Sum of squares method for each dataset
SSx1 = (sum((y1-meany1)**2))
SSy1 = (sum((x1-meanx1)**2))
SSx2 = (sum((y2-meany2)**2))
SSy2 = (sum((x2-meanx2)**2))
SSx3 = (sum((y3-meany3)**2))
SSy3 = (sum((x3-meanx3)**2))
SSx4 = (sum((y4-meany4)**2))
SSy4 = (sum((x4-meanx4)**2))
print(SSx1)
print(SSy1)
print(SSx2)
print(SSy2)
print(SSx3)
print(SSy3)
print(SSx4)
print(SSy4)

# Regression Sum of Squares method for each dataset, https://formulas.tutorvista.com/math/regression-sum-of-squares-formula.html
test1 = ((x1-meanx1)-(y1-meany1))
SSx1y1 = sum(test1)
b1 = (SSx1y1/SSx1)
b0 = (meany1 - (b1*meanx1))
yhat = (b0 + b1*(x1))

SSR = (sum((yhat-meany1)**2))
print(SSR)

# Adapted from Lectures on Simple Linear Regression by Dr. Ian McGloughlin
# Calculating the line of best fit for all four datasets of Anscombe's Quartet
# Calculate the best values for m and c.

# Line of best fit for Dataset 1 of Anscombe's Quartet
# First calculate the means of x1 and y1 for Dataset 1.
# Adapted from https://www.tutorialspoint.com/python/python_functions.htm to return multiple results from a function.
# Adapted from https://www.geeksforgeeks.org/g-fact-41-multiple-return-values-in-python/ to return multiple results from a function.
def bestfit(x, y):
    x_avg = np.mean(x)
    y_avg = np.mean(y)

    # Subtract the mean of x from all the x1 values and the mean of y from all the y1 values.
    x_zero = x - x_avg
    y_zero = y - y_avg

    # The best m is found by the following calculation.
    m = np.sum(x_zero * y_zero) / np.sum(x_zero * x_zero)
    # Use m from above to calculate the best c.
    c = y_avg - m * x_avg

    return(m, c)

# Runs the function bestfit for each dataset in the Anscombe Quartet.
line1 = bestfit(x1,y1)
line2 = bestfit(x2,y2)
line3 = bestfit(x3,y3)
line4 = bestfit(x4,y4)
m1, c1 = line1[0], line1[1]
m2, c2 = line2[0], line1[1]
m3, c3 = line3[0], line1[1]
m4, c4 = line4[0], line1[1]

# Subplot 1: Plots Dataset 1 in the Ansombe Quartet
# Creates subplots in a 2 row by 2 column format, the '1' means that the code preceeding will apply to the 1st subplot.
plt.subplot(221)
plt.plot(x1, y1, 'r.', label="Dataset I") 
plt.title("Dataset I")
plt.xlabel("X")
plt.ylabel("Y")
# Create custom limits for x-axis and y-axis - Adapted from https://python-graph-gallery.com/191-custom-axis-on-matplotlib-chart/
plt.xlim(2,20)
plt.ylim(2,15)
# Can use polyfit for bestline rather than calculate it as above in "bestfit" formula
# z1 = np.polyfit(x1, y1, 1)
# p1 = np.poly1d(z1)
# plt.plot(x1,p1(x1),"b--")
plt.plot(x1, m1 * x1 + c1, 'b-', label='Best fit line')

# Subplot 2: Plots Dataset 2 in the Ansombe Quartet
plt.subplot(222)
plt.plot(x2, y2, 'r.', label="Dataset II")
plt.title("Dataset II")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(2,20)
plt.ylim(2,15)
plt.plot(x2, m2 * x2 + c2, 'b-', label='Best fit line')

# Subplot 3: Plots Dataset 3 in the Ansombe Quartet
plt.subplot(223)
plt.plot(x3, y3, 'r.', label="Dataset III")
plt.title("Dataset III")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(2,20)
plt.ylim(2,15)
plt.plot(x3, m3 * x3 + c3, 'b-', label='Best fit line')

# Subplot 4: Plots Dataset 4 in the Ansombe Quartet
plt.subplot(224)
plt.plot(x4, y4, 'r.', label="Dataset IV")
plt.title("Dataset IV")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(2,20) # Change the limits to custom limits of 0 to 20
plt.ylim(2,15) # Change the limits to custom limits of 0 to 15
plt.plot(x3, m3 * x3 + c3, 'b-', label='Best fit line')


plt.subplots_adjust(left=0, bottom=0, right=2, top=2, wspace=0.25, hspace=0.25)

plt.show()
```
