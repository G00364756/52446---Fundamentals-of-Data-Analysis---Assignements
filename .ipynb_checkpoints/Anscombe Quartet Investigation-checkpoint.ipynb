### 2. Background of Anscombe Quartet Dataset

Frank Anscombe was an English statician, born in 1918 and died in 2001. He made numerous contributions to the field of statistics. One such contribution is a paper he published in the The American Statistician, Vol. 27, No. 1. (Feb., 1973).  Through use of 4 simple datasets which have know come to be known as "Anscombe's Quartet dataset" he was able to demonstrate the dangers of soley using summary statistics to analyse data. Summary statistics are very useful in statistical analysis but "Graphs are essential to good statistical analysis...", which is a direct quote from his 1973 paper as previously mentionsed. The Anascombe Quartet dataset, as the name suggests, consists of 4 datasets...
