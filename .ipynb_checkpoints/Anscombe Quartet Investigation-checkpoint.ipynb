### 2. Background of Anscombe Quartet Dataset

Frank Anscombe was an English statician, born in 1918. Frank Anscombe died in 2001 but throughout his life he made numerous contributions to the field of statistics. One such contribution is the use of 4 simple datasets to demonstrate the dangers of soley using summary statistics to analyse data, in a scientific paper he published in the "The American Statistician, Vol. 27, No. 1. (Feb., 1973)". These 4 datasets since come to be known as "Anscombe's Quartet dataset". In not so many words, Anscombe's lesson to the world was that, while summary statistics can be very useful in statistical analysis, they can also be misleading which is why it is so important to visualise data in tandem with summary statistics. Both will contribute to the understanding of the data. The Anascombe Quartet dataset, as the name suggests, consists of 4 datasets which can be seen in the figure below.

### 3. Interesting aspects of the Anscombe Quartet Dataset

A comparison of the summary statictics of each dataset would lead one to believe that all the datasets in Anscombe's Quartet are almost identical. Some summary statistics calculations have been conducted on the quartet to demonstrate this point below. Similar summary statistics calculations of Anscombe's 1973 paper have been used.

List of summary statistics calculations used:-

* Number of observations: $n$
* Mean of the x's: $\bar{x}$
* Mean of the y's: $\bar{y}$
* Regression coefficient(the slope): $m$
* Equation of regression line: $y =mx + c$
* Sum of squares: $(x - \bar{x})^2$

[7](http://csrgxtu.github.io/2015/03/20/Writing-Mathematic-Fomulars-in-Markdown/) See References : Writing mathematial formulas in Markdown

```python
import numpy as np
import pandas as pd
import seaborn as sns
# Load the the Anscombe's Quartet csv data set from specified folder on machine.
df = pd.read_csv("/Users/aoconnor/Desktop/Programming/Semester 2 (2018_19)/52446---Fundamentals-of-Data-Analysis---Assignments/Anscombe_Quartet_Dataset_test.csv")
# Split up the data into the Xs and the Ys for each Quartet
Quart1 = df.loc[df.loc[:, 'dataset'] == 'I']
XdatasetI = Quart1.iloc[:, 2]
YdatasetI = Quart1.iloc[:, 3]
Quart2 = df.loc[df.loc[:, 'dataset'] == 'II']
XdatasetII = Quart2.iloc[:, 2]
YdatasetII = Quart2.iloc[:, 3]
Quart3 = df.loc[df.loc[:, 'dataset'] == 'III']
XdatasetIII = Quart3.iloc[:, 2]
YdatasetIII = Quart3.iloc[:, 3]
Quart4 = df.loc[df.loc[:, 'dataset'] == 'IV']
XdatasetIV = Quart4.iloc[:, 2]
YdatasetIV = Quart4.iloc[:, 3]
```

```python
# Changes all arguments of the array to type float adapted from https://stackoverflow.com/questions/28393103/typeerror-cannot-perform-reduce-with-flexible-type
x1 = np.array((XdatasetI)).astype(np.float)
y1 = np.array(((YdatasetI))).astype(np.float)
x2 = np.array((XdatasetII)).astype(np.float)
y2 = np.array((YdatasetII)).astype(np.float)
x3 = np.array((XdatasetIII)).astype(np.float)
y3 = np.array((YdatasetIII)).astype(np.float)
x4 = np.array((XdatasetIV)).astype(np.float)
y4 = np.array((YdatasetIV)).astype(np.float)

# (n) length of the 4 Datasets
NX1 = len(x1)
NY1 = len(y1)
NX2 = len(x2)
NY2 = len(y2)
NX3 = len(x3)
NY3 = len(y3)
NX4 = len(x4)
NY4 = len(y4)
print("Number of observations for Dataset I:", "%.2f" % NY1)
print("Number of observations for Dataset II:", "%.2f" % NY2)
print("Number of observations for Dataset III:", "%.2f" % NY3)
print("Number of observations for Dataset IV:", "%.2f" % NY4)
print("\n")

# Mean of the x's and mean of the y's for each dataset
meanx1 = (sum(x1)/len(x1))
meany1 = (sum(y1)/len(y1))
meanx2 = (sum(x2)/len(x2))
meany2 = (sum(y2)/len(y2))
meanx3 = (sum(x3)/len(x3))
meany3 = (sum(y3)/len(y3))
meanx4 = (sum(x4)/len(x4))
meany4 = (sum(y4)/len(y4))

print("Mean of x values for Dataset I is:", "%.2f" % meanx1, "and the mean of y values for Dataset I is:", "%.2f" % meany1)
print("Mean of x values for Dataset II is:", "%.2f" %  meanx2, "and the mean of y values for Dataset II is:", "%.2f" % meany2)
print("Mean of x values for Dataset III is:", "%.2f" %  meanx3, "and the mean of y values for Dataset III is:", "%.2f" % meany3)
print("Mean of x values for Dataset IV is:", "%.2f" % meanx4, "and the mean of y values for Dataset IV is:", "%.2f" % meany4)
print("\n")

# Sum of squares method for each dataset
SSy1 = (sum((y1-meany1)**2))
SSx1 = (sum((x1-meanx1)**2))
SSy2 = (sum((y2-meany2)**2))
SSx2 = (sum((x2-meanx2)**2))
SSy3 = (sum((y3-meany3)**2))
SSx3 = (sum((x3-meanx3)**2))
SSy4 = (sum((y4-meany4)**2))
SSx4 = (sum((x4-meanx4)**2))
#print("Number of observations, II Dataset:", "%.2f" % SSx1)
print("The sum of squares of the x values for Dataset I is:","%.2f" % SSx1, "and the sum of squares of the y values for Dataset I is:", "%.2f" % SSy1)
print("The sum of squares of the x values for Dataset II is:","%.2f" % SSx2, "and the sum of squares of the y values for Dataset II is:", "%.2f" % SSy2)
print("The sum of squares of the x values for Dataset III is:","%.2f" % SSx3, "and the sum of squares of the y values for Dataset III is:", "%.2f" % SSy3)
print("The sum of squares of the x values for Dataset IV is:","%.2f" % SSx4, "and the sum of squares of the y values for Dataset IV is:", "%.2f" % SSy4)
print("\n")
# Adapted from Lectures on Simple Linear Regression by Dr. Ian McGloughlin
# Calculating the line of best fit for all four datasets of Anscombe's Quartet
# Calculate the best values for m and c.

# Line of best fit for Dataset 1 of Anscombe's Quartet
# First calculate the means of x1 and y1 for Dataset 1.
# Adapted from https://www.tutorialspoint.com/python/python_functions.htm to return multiple results from a function.
# Adapted from https://www.geeksforgeeks.org/g-fact-41-multiple-return-values-in-python/ to return multiple results from a function.
def bestfit(x, y):
    x_avg = np.mean(x)
    y_avg = np.mean(y)

    # Subtract the mean of x from all the x1 values and the mean of y from all the y1 values.
    x_zero = x - x_avg
    y_zero = y - y_avg

    # The best m is found by the following calculation.
    m = np.sum(x_zero * y_zero) / np.sum(x_zero * x_zero)
    # Use m from above to calculate the best c.
    c = y_avg - m * x_avg

    return(m, c)

# Runs the function bestfit for each dataset in the Anscombe Quartet.
# bestfit outputs the slope and constant for each dataset, this information will form the line of bestfit for each quartet
line1 = bestfit(x1,y1)
line2 = bestfit(x2,y2)
line3 = bestfit(x3,y3)
line4 = bestfit(x4,y4)
m1, c1 = line1[0], line1[1]
m2, c2 = line2[0], line1[1]
m3, c3 = line3[0], line1[1]
m4, c4 = line4[0], line1[1]

print("The slope (m) for Dataset I is:","%.2f" % m1, "and the constant (c) for Dataset I is:", "%.2f" % c1)
print("The slope (m) for Dataset II is:","%.2f" % m2, "and the constant (c) for Dataset II is:", "%.2f" % c2)
print("The slope (m) for Dataset III is:","%.2f" % m3, "and the constant (c) for Dataset III is:", "%.2f" % c3)
print("The slope (m) for Dataset IV is:","%.2f" % m4, "and the constant (c) for Dataset IV is:", "%.2f" % c4)

# Regression Sum of Squares method for each dataset, https://formulas.tutorvista.com/math/regression-sum-of-squares-formula.html
# test1 = ((x1-meanx1)-(y1-meany1))
# SSx1y1 = sum(test1)
# b1 = (SSx1y1/SSx1)
# b0 = (meany1 - (b1*meanx1))
# yhat = (b0 + b1*(x1))

# SSR = (sum((yhat-meany1)**2))
# print(SSR)
```

## 5. Explaination of why Anascombe's Quartet Dataset is interesting

Examination of the above graphs show that the previous assumptions that the datasets are similar in nature are incorrect. The summary statistics have mislead us to believe this. 
This can be down to a number of reasons. Linear regression modeling is the most useful form of analysis for Dataset I when observing the locations of the data points within this dataset. In contrast if we look at Dataset II linear regression modeling is not the most appropriate tool to analyse the dataset. Dataset II clearly takes the form of a polynomial function and polynomial regression analysis would be a more appropriate tool in analysing Dataset II.
For Dataset III and Dataset IV the an alternative problem exists for both, the existence of outliers. While linear regression modelling is the correct tool for analysing these 2 datasets the results of the regression modelling are being skewed by the existence of these outliers. We can observe that the trend lines in he "Dataset III" graph and the "Dataset IV" graph do  not follow the path of the majority of the data points. Suspected outliers should be analysed separately to dicipher whether they are incorrect values. If this is the case, outliers can be omitted from analysis. If this had been achieved for both graphs the linear regression analysis would be much different for the "Dataset III" and "Dataset IV" graphs. Again these conclusions are only clear to us when the dataset points have been graphed. The demonstrations in headings 3 and 4 of this jupyter notebook hold a very important message in analysing data. Reliance should not soley be placed on the use of summary statistics. Datasets should be graphed in tandem with the use of summary statistics when analysing them, to identify outliers or whether the correct mathematical techniques have been used to analyse the data. This is the key message Anascombe was demonstrating in his 1973 paper and this message holds true to this day.
